````python
import pandas as pd

DATA_PATH = '../data/stockmarket'

file = pd.read_csv(f'{DATA_PATH}/indices.csv', sep='\t', parse_dates=['Date'])

file
````

````python
df = file.pivot(index='Date', columns='Ticker', values='Price').loc['2024']

df
````

````python
import matplotlib.pyplot as plt

df.plot()
````

````python
normalized_df = df / df.iloc[0]
ax = normalized_df.plot()
````

````python
df2 = df.pct_change()

df2
````

````python
df2.plot()
````

````python
df3 = df2.corr()

df3
````

````python
from pandas.plotting import scatter_matrix
````

````python
axes = scatter_matrix(df2, diagonal='kde')

nrow, ncol = axes.shape
for i in range(nrow):
    for j in range(ncol):
        ax = axes[i, j]
    txt = f'Correlation = {df3.iloc[i, j]:.3f}'
    ax.text(0, 0, txt)
````
